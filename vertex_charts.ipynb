```python
window_sizes = [5, 10, 15, 20]
hash_modules = [128, 256, 512, 1024]
thresholds = [1, 10, 20]

tests = [(window, hash_module, threshold) 
         for window in window_sizes 
         for hash_module in hash_modules
         for threshold in thresholds]
```

```python
for window, h, t in tests:
    Logger.get_instance().set_verbosity(0)
    vertex.clustering("datasets", "prova", window_size=window, hash_module=h, threshold=1)
    print((window, h, t), evaluation_metrics.precision_recall("prediction.csv", "golden_set.csv"))
```
