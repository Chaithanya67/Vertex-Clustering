# Test parametri
In questa sezione si trovano i vari test effettuati su vertex

### Precision, recall, f1

```python
#directory e ground truth
website = "study"
ground_truth_file = "datasets/study_ground_truth.csv"
```

Per testare singolo parametro basta porre gli altri ad un singlo valore

```python
#definizione parametri
window_sizes = [10]
hash_modules = [64, 128, 256, 512, 1024, 2048]
thresholds = [26]

tests = [(window, hash_module, threshold) 
         for window in window_sizes 
         for hash_module in hash_modules
         for threshold in thresholds]
```

```python
#esecuzione dei test
def run_tests(tests):
    results = dict()
    for window, h, t in tests:
        print("Starting evaluating with", window, h, t)
        Logger.get_instance().set_verbosity(0) # non strettamente necessario
        vertex.clustering("datasets", website, window_size=window, hash_module=h, threshold=t)
        results[(window, h, t)] = evaluation_metrics.precision_recall("prediction.csv", ground_truth_file)
        print("Evaluated with:", results[(window, h, t)])
    return results
```

```python
results = run_tests(tests)
```

```python
def get_results_ordered(results):
    result_ordered = sorted(results.items(), key=lambda x: x[1][2], reverse=True)
    return result_ordered

def get_best_params(results):
    return get_results_ordered(results)[0][0]

def save_results_on_file(results):
    import time
    time_id = time.strftime("_%Y_%m_%d_%H_%M_%S") # per evitare di sovrascrivere i vecchi test

    result_ordered = get_results_ordered(results)
    test_filename = "test_result/" + website + time_id + ".txt"
    file = open(test_filename, "w", encoding="utf-8")

    for test, performance in result_ordered:
        window, h, t = test
        precision, recall, f1 = performance
        record = "window: " + str(window) + " hash_module: " + str(h) + " threshold: " + str(t)
        record += " precision: " + str(precision) + " recall: " + str(recall) + " f1: " + str(f1) 
        file.write(record + "\n")
    file.close()
```

```python
save_results_on_file(results)
get_best_params(results)
```

### Grafici

```python
def get_metric_index_from_name(metric_name):
    metric_to_index = {
        "recall" : 0,
        "precision" : 1,
        "f1" : 2
    }
    return metric_to_index[metric_name.lower()]
    
def get_feature_index_from_name(feature_name):    
    feature_to_index = {
        "window-size" : 0,
        "hash-module" : 1,
        "threshold" : 2
    }
    return feature_to_index[feature_name.lower()]
    
def extract_feature(results, feature_name, metric_name):
    index_feature = get_feature_index_from_name(feature_name)
    index_metric = get_metric_index_from_name(metric_name)
    results = get_results_ordered(results) 
    feature, feature_results = [], []
    for test, performance in results:
        feature.append(test[index_feature])
        feature_results.append(performance[index_metric])
    return feature, feature_results
```

```python
def plot_results_on_feature(results, feature, metric_name):
    feature.sort(), results.sort()
    plt.plot(feature, results, label=metric_name)
    plt.legend()

def plot_metric_on_feature(results, feature_name, metric_name):
    feature, results = extract_feature(results, feature_name, metric_name)
    plot_results_on_feature(results, feature, metric_name)
    
def plot_all_on_feature(results, feature_name, label):
    plot_metric_on_feature(results, feature_name, "Recall")
    plot_metric_on_feature(results, feature_name, "Precision")
    plot_metric_on_feature(results, feature_name, "F1")
    plt.ylabel("Percentage")
    plt.xlabel(label)
```

```python
plot_all_on_feature(results, "hash-module", "Hash module size")
```

## Benchmark

### Complessità temporale

```python
number_of_tests = 30
input_size_tests = [ 100 * (i+1) for i in range(0, number_of_tests)]
```

```python
import time

measured_times = []

for input_size in input_size_tests:
    print("Starting with input size:", input_size, "...........", end=" ")
    Logger.get_instance().set_verbosity(0)
    start_time = time.time()
    vertex.clustering("datasets", website, input_limit = input_size)
    end_time = time.time()
    time_elapsed = end_time - start_time
    measured_times.append( time_elapsed )
    print("Test ended")
```

```python
measured_times
```

```python
def plot_time_on_input_size(input_size, times):
    plt.plot(input_size, times)
    plt.xlabel("Input size")
    plt.ylabel("Seconds")
    
plot_time_on_input_size(input_size_tests, measured_times)
```
